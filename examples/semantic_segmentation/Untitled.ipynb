```python
import labelme
import numpy as np
import PIL.Image
import json

img_file = 'data_annotated/2011_000003.jpg'
data_file = 'data_annotated/2011_000003.json'
out_img_file = 'test_out'
label_file = 'labels.txt'

class_names = []
class_name_to_id = {}
```

```python
for i, line in enumerate(open(label_file).readlines()):
    class_id = i-1
    class_name = line.strip()
    class_name_to_id[class_name] = class_id
    if class_id == -1:
        assert class_name == '__ignore__'
        continue
    elif class_id == 0:
        assert class_name == '_background_'
    class_names.append(class_name)
```

```python
print(class_names)
print(class_name_to_id)
```

```python
img = np.asarray(PIL.Image.open(img_file))
data = json.load(open(data_file))
lbl = labelme.utils.shapes_to_label(img_shape=img.shape,
                                    shapes=data['shapes'],
                                    label_name_to_value=class_name_to_id)
```

```python
colormap = labelme.utils.label_colormap(255)
viz = labelme.utils.draw_label(lbl, img, class_names, colormap=colormap)
```

```python
PIL.Image.fromarray(viz)
```
